<h3 style="color:gray">Q1. What are the airlines in the dataset , accompanied by their frequency</h3>

```python
# check how many airlines are there in the data set
data['airline'].nunique()
```

```python
#showing the names of the airlines in the dataset
data['airline'].unique()
```

```python
#frequencies of each airline
data['airline'].value_counts()
```

```python
#Showing all the airlines with their number of flights in bar graph
data['airline'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])
plt.title("Airlines with frequencies")
plt.ylabel("Airline")
plt.xlabel("Number of flights")
plt.show()
```

<h3 style="color:gray">Q.2 Show Bar Graphs representing the Departure Time and Arrival Time</h3>

```python
data.head()
```

```python
#Showing frequency of the departure time for all the flights 
data['departure_time'].value_counts()
```

```python
#Showing frequency of the arrival time for all the flights 
data['arrival_time'].value_counts()
```

```python
plt.figure(figsize = (16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index , data['departure_time'].value_counts().values , color=['r','b'])
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index , data['arrival_time'].value_counts().values , color=['g','y'])
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")


plt.show()
```

<h3 style="color:gray">Q3. Show Bar Graphs representing the Souce City And Destination City</h3>

```python
data.head()
```

```python
#Showing the source city of the flights
data['source_city'].value_counts()
```

```python
#Showing the destination city of the flights
data['destination_city'].value_counts()
```

```python
# Showing the source city and the destination city for the flights with count 
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.barh(data['source_city'].value_counts().index , data['source_city'].value_counts().values , color=['m','g'])
plt.title("Source City wit No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.subplot(1,2,2)
plt.barh(data['destination_city'].value_counts().index , data['destination_city'].value_counts().values , color=['y','r'])
plt.title("Destination City wit No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")
```
