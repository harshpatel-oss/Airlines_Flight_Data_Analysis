<h3 style="color:gray">Q10. List all the fastests business flights from Delhi to Mumbai.</h3>

```python
data.head()
```

```python
minDuration= data[
    (data['source_city'] == 'Delhi') &
    (data['destination_city'] == 'Mumbai') &
    (data['class'] == 'Business')
]['duration'].min()
minDuration
```

```python
minDurationBusinessFlightsFromDelhiToMumbai = data[
    (data['source_city'] == 'Delhi') &
    (data['destination_city'] == 'Mumbai') &
    (data['class'] == 'Business')&
    (data['duration']==minDuration)
]
minDurationBusinessFlightsFromDelhiToMumbai
```
